## Import packages

```python
import advertools as adv
import adviz
import pandas as pd
```

## Define main variables

```python
url = "https://www.anthropic.com"
output_file = "anthropic.jsonl"
```

## Crawl a website

```python
adv.crawl(url_list=url, output_file=output_file, follow_links=True)
```

## Read crawl DataFrame and display

```python
crawldf = pd.read_json(output_file, lines=True)
print(crawldf.shape)
crawldf
```

## Fetch XML sitemap

```python
sitemap = adv.sitemap_to_df(url + "/robots.txt")
sitemap
```

## Crawled but not in sitemap

```python
set(crawldf['url']).difference(sitemap['loc'])
```

## In sitemap but not crawled

```python
set(sitemap['loc']).difference(crawldf['url'])
```

## Create a function for getting crawled URLs not found in the sitemap

```python
def crawled_not_in_sitemap(url, output_file):
    adv.crawl(url, output_file, follow_links=True)
    sitemap = adv.sitemap_to_df(url + "/robots.txt")
    return set(crawldf['url']).difference(sitemap['loc'])
```

## Analyze links on the crawl DataFrame

```python
links_df = adv.crawlytics.links(crawldf, internal_url_regex="www.anthropic.com")
links_df
```

## Get external links and count top linked domains

```python
external_links = links_df[~links_df['internal']]['link'].drop_duplicates()
adviz.value_counts(adv.url_to_df(external_links.dropna())['netloc'], title="Top linked domains from <b>anthropic.com</b>")
```

## Run AI workflows in bulk

A simple for-loop to send body text content to an LLM (ChatGPT) and get their summaries

```python
import os
from openai import OpenAI
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
```

```python
crawldf[['url', 'title', 'body_text']]
```

```python
summaries = []

for page in crawldf["body_text"]:
    resp = client.chat.completions.create(
        model="gpt-4o",
        messages=[{
            "role": "user",
            "content": f"""please summarize the following text
            using between one hundred and two hundred words,
            respond with the summary text only:

            {page}
            """
        }])
    summaries.append(resp)
```

```python
crawldf[['url', 'title']].head(5).copy().assign(summary=[summary.model_dump()['choices'][0]['message']['content'] for summary in  summaries]) 
```
